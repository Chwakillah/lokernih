```python
# 2. Bersihkan nama kolom dari spasi
df.columns = df.columns.str.strip()

# 3. Ambil kolom yang diinginkan
df = df[['Perusahaan', 'Posisi', 'Lokasi', 'Gaji', 'Sistem Kerja', 'Pendidikan Minimal', 'Pengalaman Kerja', 'Kriteria']]

# 4. Tampilkan
df.head()
```

```python
import re
import string
import nltk

# Fungsi untuk menghapus emoji
def remove_emoji(tweet):
    if tweet is not None and isinstance(tweet, str):
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F700-\U0001F77F"  # alchemical symbols
            u"\U0001F780-\U0001F7FF"  # geometric shapes extended
            u"\U0001F800-\U0001F8FF"  # supplemental arrows-C
            u"\U0001F900-\U0001F9FF"  # supplemental symbols and pictographs
            u"\U0001FA00-\U0001FA6F"  # chess symbols
            u"\U0001FA70-\U0001FAFF"  # symbols and pictographs extended-A
            u"\U0001F004-\U0001F0CF"  # additional emoticons
            u"\U0001F1E0-\U0001F1FF"  # flags
            "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', tweet)
    return tweet

# Fungsi untuk menghapus simbol
def remove_symbols(tweet):
    if tweet is not None and isinstance(tweet, str):
        return re.sub(r'[^a-zA-Z0-9\s]', '', tweet)  # Menghapus semua simbol
    return tweet

# Fungsi untuk menghapus angka
def remove_numbers(tweet):
    if tweet is not None and isinstance(tweet, str):
        return re.sub(r'\d', '', tweet)  # Menghapus semua angka
    return tweet

# Fungsi untuk menghapus username
def remove_username(text):
    return re.sub(r'@[^\s]+', '', text)

# Proses pembersihan data
df['Kriteria_Clean'] = df['Kriteria'].apply(remove_username)
df['Kriteria_Clean'] = df['Kriteria_Clean'].apply(remove_emoji)
df['Kriteria_Clean'] = df['Kriteria_Clean'].apply(remove_symbols)
df['Kriteria_Clean'] = df['Kriteria_Clean'].apply(remove_numbers)

df
```

```python
#Ubah ke huruf kecil
def case_folding(text):
    if isinstance(text, str):
        lowercase_text = text.lower()
        return lowercase_text
    else:
        return text

df['Kriteria_CF'] = df['Kriteria_Clean'].apply(case_folding)
df.head(10)
```

```python
# Tokenize teks dari
df['Kriteria_Token'] = df['Kriteria_CF'].fillna('').apply(lambda x: x.split())
df
```

```python
#stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_text(text):
  return [stemmer.stem(word) for word in text]

df['Kriteria_Stemming'] = df['Kriteria_Token'].apply(stem_text)
df.head(40)
```

```python
#stopwords
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

# Line 2: Ambil daftar stopwords bahasa Indonesia
stopword_list = set(stopwords.words('indonesian'))
```

```python
def remove_stopwords(text):
    return [word for word in text if word not in stopword_list]

df['Kriteria_Stopword'] = df['Kriteria_Stemming'].apply(remove_stopwords)

# Line 6 (Opsional): Jika kamu ingin hasil berupa string, bukan list
df['Kriteria_Stopword'] = df['Kriteria_Stemming'].apply(lambda x: ' '.join(remove_stopwords(x)))

# Line 7: Tampilkan dataframe
df
```
